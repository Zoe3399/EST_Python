### 지수와 제곱근

```python
import sympy
print(sympy.__version__)
```

```python
import sympy as sp
from sympy import symbols, expand, factor, solve

x = sp.Symbol('x')
expr = x**2 + 2*x + 1

print("전개:", sp.expand(expr))
print("인수분해:", sp.factor(expr))
print("해:", sp.solve(expr, x))
```

```python
import sympy as sp
from sympy import symbols, expand, factor, solve
k = sp.Symbol('k')

num1 = sp.solve(4 - k + 2)
num1
```

```python
equation1 = k -2 - 4
equation2 = 2 * k - 10
equation3 = 8 - k/2

print("방정식1:", sp.solve(equation1))
print("방정식2:", sp.solve(equation2))
print("방정식3:", sp.solve(equation3))
```

다음 중 **방정식과 항등식**을 모두 고르세요.

(1) x + x = 2x    <br>  >> 좌변의 x+x = 2x가 되므로 우변 2x와 같이 **항등식**임. <br>
(2) 2x + 1 < 6    <br>  >> 등호가 없기 때문에 등식이 아님 <br>
(3) 2x - x = x    <br>  >> 좌변의 2x - x = x가 되므로 우변 x와 같이 **항등식**임. <br>
(4) 2 + 5 = 7     <br>  >> 2+5 = 7에서 미지수가 없기 때문에 방정식도 아니고 항등식도 아님. <br>
