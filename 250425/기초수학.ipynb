```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X 값을 양수로 설정 (-1에서 1로)
x = np.linspace(-1, 1, 100)
```

```python
# y 값을 계산
y = 2 * x + 1
```

```python
# 그래프를 그리기
plt.plot(x, y)
```

```python
# x축과 y축의 범위를 설정
plt.xlim(-1, 1) # x축은 -1에서 1까지
plt.ylim(-1, 3) # y축은 -1에서 3까지(y = 2*1+1 = 3)
```

```python
# 원점(0,0)을 포함한 1사분면에 맞는 축 설정
plt.axhline(0, color='black', lw=0.5, ls='--')  # x축
plt.axvline(0, color='black', lw=0.5, ls='--')  # y축
```

```python
#제목과 레이블 추가
plt.title('Line Graph in the First Quadrant')
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
# 그리드 추가
plt.grid()
```

```python
# 그래프 보여주기
plt.show()
```
